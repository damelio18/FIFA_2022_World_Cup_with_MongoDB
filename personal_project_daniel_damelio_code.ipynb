## Extract 

In order to access the event data for the FIFA World Cup 2022, I need to acquire the unique match ids for each match at the tournament. However, to acquire the match ids, I first need to know the tournament's unique competition and season ids. This can be achieved by the following steps:

- __Step 1__: Access the Statsbomb Competitions API to find the unique competition and season ids for the FIFA World Cup 2022.
<br>

- __Step 2__: Access the Statsbomb Matches API using the unique competition and season ids obtained from Step 1. Then extract the unique match ids for the tournament.
<br>

- __Step 3__: Extract the data from the Statsbomb Events API using the unique match ids for the FIFA World Cup 2022 from Step 2.
<br>


### Step 1: Get Competition and Season ids

The code below shows the process of filtering the Statsbomb Competitions API to obtain the unique competition and season ids for the FIFA World Cup 2022. 

```python
# Search criteria
competition_name = 'FIFA World Cup'
season_name = '2022'

# Access API and save as dictionary object
comps_dict = sb.competitions(fmt = "dict")

# Filter dictionary based on search criteria
comps_dict = {k: v for k, v in comps_dict.items() if v['competition_name'] == competition_name 
                                                  and v['season_name'] == season_name
              }

# Obtain season and competition ids
competition_id = comps_dict[(list(comps_dict.keys())[0])]['competition_id']
season_id = comps_dict[(list(comps_dict.keys())[0])]['season_id']

# Print results
print('FIFA World Cup 2022 item in Statsbomb Competitions API:\n')
pprint(list(comps_dict.values()))
print('-' * 30)
print(f'Competition ID for FIFA World Cup 2022 = {competition_id}')
print(f'Season ID for FIFA World Cup 2022 = {season_id}')
```

### Step 2: Get Match ids

Using the unique ids obtained in Step 1, Step 2 extracts the unique match ids for the FIFA World Cup 2022 and saves them into a list object.

```python
# Access API and save as dictionary object
matches_dict = sb.matches(competition_id = competition_id, season_id = season_id, 
                          fmt = "dict")

# Get match id's available from the Matches API
match_ids = list(set([matches_dict[key]['match_id'] for key in matches_dict]))

# Print results
print(f'Number of matches available = {len(matches_dict)}')
print('-' * 30)
print(f'First 5 match ids for FIFA World Cup 2022: {match_ids[:5]}...')
```

### Step 3: Extract Events

The final stage of the extraction iterates through the list of match ids from Step 2 and saves the events to both tuple and dictionary objects.

```python
# Empty tuple and dictionary
events_tuple = (list())
events_dict = {}

# Loop through match ids
for match in match_ids:
    
    # Extract events data for match
    match_events = sb.events(match_id = match, fmt = "dict")
    
    # Save events to tuple and dictionary
    events_tuple.append(match_events)
    events_dict.update(match_events)

# Print results   
print(f'Matches extracted = {len(events_tuple)}')
print('-' * 30)
print(f'Events extracted = {len(events_dict):,}')
print('-' * 30)
print(f'Size events object = {round(getsizeof(events_dict) * 0.000001,2)} mb')
```

### Load Events to Mongo DB

Loading the 234,602 events in one attempt using the '.insert_many()' function proved problematic and returned an 'AutoReconnect' error. After investigating, I found that the load I was attempting exceeded the 'maxWriteBatchSize' for the Mongo DB cluster. 

The solution was to iterate through each element of the tuple created in Section 3.3. The tuple contains 64 elements (one for each match), and within each element, all the events for each match are contained (approximately 3,700 events per match). This solution broke down the workload into more manageable batches for the load to the Mongo DB cluster.

```python
counter = 0

# Loop through each match
for match in events_tuple:
    
    # Insert all events from match into the collection
    events_coll.insert_many([value for key, value in match.items()])

    # Print status of loading
    counter += 1
    if (counter % 10) == 0:
        print(f'Events from {counter} matches loaded')

# Print results    
print(f'Loading Completed: {counter} matches loaded')
print('-' * 30)
print(f'Number of events loaded as documents = {events_coll.count_documents({}):,}')
```

## Transformation

### Inspection

There are 33 different types of events in the collection, and the structure of each document in the collection is different depending on the type of event. Below is an example data structure for a successful penalty event, which is now stored as a document in the Mongo DB collection.

If the event were a pass, the 'shot' key would be replaced with a 'pass' key, which will store a different number of keys and values relating to the pass. These changes in the data structure make Mongo DB an appropriate choice to store the data compared to more rigid Relational Database Management System (RDBMS) alternatives.

The data structure is already embedded in a clean, logical structure. Whilst there is no right or wrong answer for how to model the data in Mongo DB, I preferred to keep the data as embedded as it is a more efficient way to store the data, especially for data that are regularly accessed together.

```python
# Filter: goals and penalties (not case sensitive)
query = {
    "shot.outcome.name": {"$regex": "Goal", "$options": "i"},
    "shot.type.name": {"$regex": "Penalty", "$options": "i"},
        }

# View one document
limit = 1

# Execute query
cursor = events_coll.find(filter = query,
                          limit = limit)

# Print results
print(f'Types of event in collection = {len(events_coll.distinct("type.id"))}')
print('-' * 30)

print('Sample structure for shots:\n')

try:
    for doc in cursor:
        pprint(doc)
finally:
    cursor.close()
    
del [query, limit, cursor]
```

### Check Data Quality

A benefit of working with Statsbomb data is that the data is of very high quality, and the granularity is excellent. As a result, the data extracted from the API doesn't require much, if any, cleaning. However, it is still necessary to do some basic data quality checks.

Firstly, I check for missing values. Due to the different structures of the documents in the collection, two stages are required. The first stage is to use an aggregation pipeline to get all the unique keys in the collection. The second stage uses a nested for loop that searches for each key in each document and prints any missing values. Below, the output confirms that there are no missing values in the collection.

```python
# Pipeline to get all unique keys in the collection
pipeline = [
    {"$project": {"keys": {"$objectToArray": "$$ROOT"}}},
    {"$unwind": "$keys"},
    {"$group": {"_id": None, "uniqueKeys": {"$addToSet": "$keys.k"}}}
]

# Aggregate pipeline
result = events_coll.aggregate(pipeline)

# Save to list
unique_keys = result.next()["uniqueKeys"]

# Print results
for key in unique_keys:
    for doc in events_coll.find({key: {"$type": 10}}):
        print(f"Document with {key} missing: {doc['_id']}")
        
print('Missing values checked')
```

Secondly, duplicate values need to be checked. The output below searches for any duplicated values for the event id key, and confirms none are present in the collection.

```python
# Pipeline to search for duplicate event id's
pipeline = [
    {"$group": {"_id": "$id", "count": {"$sum": 1}}},
    {"$match": {"count": {"$gt": 1}}}
]

# Aggregate pipeline
result = events_coll.aggregate(pipeline)

# Print results
for doc in result:
    print(doc)
    
print('Duplicate events checked')
```

### Transform

Every event recorded in the API has x and y coordinates attributed to it and are stored as float values in a nested array for the 'location' key. The Statsbomb pitch coordinates can be seen in Fig. 2 below, with the units measured in yards.

I have decided to make two transformations using the location values for each event and write the changes to every document in the Mongo DB collection.

<img src = "Images/pitch_coordinates.png" width = "400" align = "center">
<font size="2"><em><center> <a href="https://coolbluedata.com/wp-content/uploads/2021/01/football_001.png" target="_blank">Fig.2 - Statsbomb Pitch Coordinates</a></center></em></font>

Firstly, I want to calculate the Euclidean distance from the goal for each event, which will benefit analysis. For example, with this calculated distance, you can easily see the distance a player is regularly shooting from and compare this with different players. Or whether a player makes crucial successful last-ditch tackles close to their goal. Secondly, I want the calculated distance converted from yards to metres, as this is a more interpretable unit.

Below is the code used to create this new calculated field and apply the changes to all documents in the collection. The first output shows the execution of the transformation, and the second output, confirms the new field has been added. In the output, the distance to goal equals 10.97 metres, which is the correct distance for a penalty kick.

```python
# Create new calculated field
events_coll.aggregate([{"$addFields": {
      "dist_to_goal": {
        "$sqrt": {
          "$sum": [
            { "$pow": [ { "$subtract": [ { "$arrayElemAt": [ "$location", 0 ] }, 120 ] }, 2 ] },
            { "$pow": [ { "$subtract": [ { "$arrayElemAt": [ "$location", 1 ] }, 40 ] }, 2 ] }
          ]
        }
      }
    }
  },
  {
    "$out": "fifa_wc_2022_events"
  }
])

# Conversion factor
yards_to_meters = 0.9144

# Update all documents 
events_coll.update_many({}, {'$mul': {'dist_to_goal': yards_to_meters}})
```

### Goalscoring Reliance

The first analysis looks at goals scored by each team at the tournament and then at Argentina and France's reliance on Messi and Mbappe's goals. The aggregation pipeline below sum's the total number of goals scored by each team at the tournament and is visualised in a bar chart.

```python
# Filter: goals, excl. penalty shootouts
match = {"$match": {
            "shot.outcome.name": "Goal",
            "period": {"$lte": 4}
            }
        }

# Group by: team, sum of goals
group = {"$group": {
            "_id": "$team.name",
            "Goals": { "$sum": 1 }
            }
        }

# Sort by: goals
sort = {"$sort": {
            "Goals": -1
            }
       }

# Choose: team, goals
project = {"$project": {
                "_id": 0,
                "Team": "$_id",
                "Goals": 1, 
            }
        }

# Create pipeline
pipeline = [match, group, sort, project]

# Aggregate pipeline
cursor = events_coll.aggregate(pipeline)

# Save as pandas dataframe
df_team_goals = pd.DataFrame(cursor)

# Set theme and window size
sns.set_theme(palette = sns.color_palette(), style = 'white',  
              rc = {'figure.figsize':(15,5)})

# Generate barplot
ax = sns.barplot(x = "Team", y = "Goals", data = df_team_goals, 
                 palette = "OrRd_r") 

# Customize the plot
plt.xticks(rotation = 45)
ax.set_title('Goals Scored per Team in 2022 FIFA World Cup', fontsize = 20, 
             fontfamily = 'monospace')
ax.set_xlabel('Team', fontsize = 10, fontfamily = 'monospace')
ax.set_ylabel('Goals Scored', fontsize = 10, fontfamily = 'monospace')
ax.tick_params(axis = 'x', rotation = 90)
sns.despine(left = True) 
plt.xticks(fontname = 'monospace', fontsize = 10)
plt.yticks(fontname = 'monospace', fontsize = 10, color = 'white')
plt.suptitle('Fig.4 - Goals Scored per Team', fontsize = 12,
             style = 'italic', y = -0.23)

# Add labels
for i in ax.containers:
    ax.bar_label(i,fontsize = 8)
```

The following aggregation pipeline examines Messi and Mbappe's goal contributions to their teams. As shown in the resulting data frame and bar chart, both players contributed significantly to their teams, with Messi scoring 46.7% of Argentina's goals and Mbappe scoring 50% of France's goals.

```python
# Filter: shots, goals, excl. penalty shootouts, team
match = {"$match": {
            "shot": {"$exists": True},
            "shot.outcome.name": "Goal",
            "period": {"$lte": 4},
            "team.name": {"$in": ["Argentina", "France"]}
            }
        }

# Group by: team name, goals, Messi goals, Mbappe goals
group = {"$group": {
            "_id": "$team.name",
            "Total_Goals": {"$sum": 1},
            "Mbappe_Goals": {"$sum": {"$cond": [{"$eq": 
                                                 ["$player.id", 3009]}, 1, 0]}},
            "Messi_Goals": {"$sum": {"$cond": [{"$eq": 
                                                ["$player.id", 5503]}, 1, 0]}}
            }
        }

# Add field: Messi and Mbappe goals
add_field_1 = {"$addFields": {
                "Messi_Mbappe_Goals": {"$add": ["$Mbappe_Goals", "$Messi_Goals"]}
                }
            }

# Add field: % of goals scored by Messi and Mbappe 
add_field_2 = {"$addFields": {
                    "Player_Reliance": {
                        "$multiply": [
                            {"$divide": [
                                {"$add": ["$Mbappe_Goals", "$Messi_Goals"]},
                                    "$Total_Goals"
                        ]
                    },
                 100
                ]
            }
        }
    }

# Choose: team, Messi and Mbappe, other goals, player reliance
project = {"$project": {
               "_id": 0,
               "Team_Name": "$_id",
               "Messi_Mbappe_Goals": "$Messi_Mbappe_Goals",
               "Other_Goals": {"$subtract": ["$Total_Goals", 
                                                {"$add": ["$Mbappe_Goals", 
                                                          "$Messi_Goals"]}]},
               "Player_Reliance": "$Player_Reliance"
            }
        }

# Create pipeline
pipeline = [match, group, add_field_1, add_field_2, project]

# Aggregate pipeline
cursor = events_coll.aggregate(pipeline)

# Save as pandas dataframe
df_goals_rel = pd.DataFrame(cursor)

# View data frame
df_goals_rel
```

### Goalscoring Efficiency

Having seen that the teams were reliant on the goals scored by their star players, we now look at the efficiency of the players in front of the goal. Expected Goals (xG) is a metric designed to measure the probability of a shot resulting in a goal. Every shot event in the Mongo DB collection has an XG value that is scaled between 0 (not likely to score) to 1 (very likely to score). 
Therefore, if a player's goal total is higher than their XG, the player is performing well and is efficient in front of the goal.

The aggregation pipeline below calculates both players' XG, goals scored, and average shot distance. The resulting data frame column 'XG_Goal_Diff' shows that Mbappe far outperformed Messi regarding efficiency in front of the goal. Meaning Mbappe was converting more chances than he was expected to.

```python
# Filter: shots, Messi or Mbapp, excl. penalty shootouts
match = {"$match": {
            "shot": {"$exists": True},
            "player.name" : {"$regex": "(Messi|Mbapp)", "$options": "i"},                                                      
            "period": {"$lte": 4}
            }
        }

# Choose: player id, player name, shot outcome, XG, distance
project = {"$project": {
                "_id": 0,
                "Player_ID" : "$player.id", 
                "Player_Name" : "$player.name",
                "Shot_Outcome" : "$shot.outcome.name",
                "XG" : "$shot.statsbomb_xg",
                "Distance" : "$dist_to_goal"
            }
        }

# Group by: player name and id, sum of XG, sum of goals, avg. distance
group = {"$group": {
            "_id": {"Player_Name" : "$Player_Name", "Player_ID" : "$Player_ID"},
            "Total_XG": {"$sum": "$XG"},
            "Goals_Scored": {"$sum": {"$cond": [ {"$eq": ["$Shot_Outcome", "Goal"]}, 1, 0 ]}},
            "Avg_Distance" : {"$avg": "$Distance"}
            }
        }

# Choose: player id, player name, XG, goals, goals minus xg, distance
project2 = {"$project": {
                "_id": 0,
                "Player_ID" : "$_id.Player_ID",
                "Player_Name" : "$_id.Player_Name",
                "XG" : "$Total_XG",
                "Goals_Scored" : "$Goals_Scored",
                "XG_Goal_Diff": {"$subtract": ["$Goals_Scored", "$Total_XG"]},
                "Avg_Distance" : "$Avg_Distance"
            }
        }

# Sort by: goals ascending
sort = {"$sort": {
            "Goals_Scored": -1
            }
       }

# Create pipeline
pipeline = [match, project, group, project2, sort]

# Aggregate pipeline
cursor = events_coll.aggregate(pipeline)

# Save as pandas dataframe
df_goals_agg = pd.DataFrame(cursor)

# View dataframe
df_goals_agg
```

### Postitioning

The final analysis shows where the players were receiving and making passes in the tournament in the form of a heatmap. Seeing where the players were involved with passes shows their involvement in their team's play. In the visual below, we can see that Mbappe was an out-and-out left winger who was attack-minded. Messi was more involved in the build-up play for the team and was roaming around the pitch more in a free role.

```python
# Filter: passes, either made or received by Messi or Mbapp
match = {"$match": {
            "pass": {"$exists": True}, # Filter for all passes
            "$or": [
                {"player.name" : {"$regex": "(Messi|Mbapp)", "$options": "i"}},
                {"pass.recipient.name": {"$regex": "(Messi|Mbapp)", "$options": "i"}}
                   ]
                    }}

# Choose: player id, player name, pass recipient id, pass recipient name, location x, location y
project = {"$project": {
                "_id": 0,
                "player_id": "$player.id",
                "player_name": "$player.name",
                "pass_recipient_id": "$pass.recipient.id",
                "pass_recipient_name": "$pass.recipient.name",
                "location_x": {"$arrayElemAt": ["$location", 0]},
                "location_y": {"$arrayElemAt": ["$location", 1]}
                        }}
# Create pipeline
pipeline = [match, project]

# Aggregate pipeline
cursor = events_coll.aggregate(pipeline)

# Save as pandas dataframe
df_passes_2 = pd.DataFrame(list(cursor))

# Filter dataframes
df_messi_pass = df_passes_2.loc[(df_passes_2['player_id'] == 5503) | (df_passes_2['pass_recipient_id'] == 5503)]
df_mbappe_pass = df_passes_2.loc[(df_passes_2['player_id'] == 3009) | (df_passes_2['pass_recipient_id'] == 3009)]
```
